```python
def calculate_efficiency(actual_output, target_output):
    """
    Returns efficiency as a percentage.
    """
    efficiency = (actual_output / target_output) * 100
    return efficiency
```

# Week 1 Operations Analyzer

## Objective
This notebook demonstrates how operational data can be transformed into useful insights using Python.

We will:
- Calculate operational efficiency.
- Classify operational performance.
- Store operational site data in dictionaries.
- Loop through each site to generate a report.

```python
def get_operational_status(efficiency):
    """
    Returns operational status based on efficiency.
    """
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:
        return "Warning"
    else:
        return "Normal"
```

## KPI Function

The function above calculates operational efficiency.

Formula:

Efficiency (%) = (Actual Output ÷ Target Output) × 100

This allows us to compare different operational sites using the same performance metric.

```python
sites = {
    "Site 1": {
        "name": "Nairobi Fuel Depot",
        "actual_output": 920,
        "target_output": 1000
    },
    "Site 2": {
        "name": "Mombasa Fuel Depot",
        "actual_output": 760,
        "target_output": 1000
    },
    "Site 3": {
        "name": "Kisumu Fuel Depot",
        "actual_output": 610,
        "target_output": 1000
    }
}
```

## Status Classification

After calculating efficiency, we classify each site:

- Below 70% → Critical
- 70% to 89% → Warning
- 90% or higher → Normal

This makes it easier for managers to quickly identify which sites need attention.

```python
print("OPERATIONS PERFORMANCE REPORT")
print("-" * 40)

for site in sites.values():
    efficiency = calculate_efficiency(
        site["actual_output"],
        site["target_output"]
    )

    status = get_operational_status(efficiency)

    print(f"Site: {site['name']}")
    print(f"Actual Output : {site['actual_output']}")
    print(f"Target Output : {site['target_output']}")
    print(f"Efficiency    : {efficiency:.2f}%")
    print(f"Status        : {status}")
    print("-" * 40)
```

## Data Model

The operational data is stored in a Python dictionary.

Each site contains:
- Name
- Actual Output
- Target Output

Using dictionaries makes it easy to organize and retrieve operational information.
